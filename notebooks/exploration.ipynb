## Selección de variables
Dado que la naturaleza de todas las variables son cualitativas, vamos a utilizar un método de tratamiento de features para este tipo de variables, que es por medio del Weight of Evidence (WOE) y el Information Value(IV), en términos de palabras lo que nos dice el WOE es que tan probable se presente el valor de nuestro target de acuerdo a la categoría de la variable que estamos evaluando, su formula se cálcula cómo 
$$WOE = ln\Big(\frac{\%\text{ de eventos positivos}}{\% \text{ de eventos negativos}}\Big) $$

Que para entender su cómo funciona esta médida veamos un ejemplo, supongamos que tenemos la siguiente tabla de frecuencias

|Variable|0|1|
|---|---|---|
|Categoría A|4|5|
|Categoría B|7|3|
Total|11|12|

Cuando queremos cálcular el WOE, este se realiza para cada categoría, así tendriamos lo siguiente:
- WOE de Categoria A = $ln(\frac{5/8}{4/11}) = ln(0.625) - ln(0.36) = 0.55$
- WOE de Categoria B= $ln(\frac{3/8}{7/11}) = ln(0.375) - ln(0.63) = -0.58$

Lo que esto nos dice es que es más probable que en caso de tener un registro que pertenezca a la Categoría A, este será positivo o 1 en este caso, de modo análogo, tenemos que si un registro pertenece a la categoría B es muy probable que este termine siento negativo.

Teniendo esto en cuenta, podemos pasar a definir lo que es el Information Value(IV), este termino nos va a indicar que cantidad de información aporta una variable respecto a la variable target u objetivo, esto se cálcula cómo:
$$IV = \sum (\%\text{ de eventos positivos} - \%\text{ de eventos positivos}) * WOE$$

Si continuamos con nuestro ejemplo, tendriamos que para la variable analizada tendríamos el siguiente information value: 
- $IV = (0.625-0.36)*0.55 + (0.375-0.63)*(-0.58) = 0.293$

Ahora, según el valor que tenemos sobre el information value, este nos va a decir que tan buena es una variable como predictora de nuestro target por medio de la siguiente referencia de valores:

|Information Value|Predictividad de la variable|
|---|---|
|Menor a 0.02|Variable que no es util|
|0.02 to 0.1|Variable con una predicción débil|
|0.1 to 0.3|Variable con una predicción media|
|0.3 to 0.5|Variable excelente para realizar una predicción|
|>0.5|Variable con un comportamiento sospechoso|

Así que según esta información, para el ejemplo, tendriamos una variable que se ajusta de forma media a nuestro objetivo de predecir nuestro target. Con esto en cuenta, vamos a realizar el cálculo del information value para cada una de las variables de nuestro conjunto de datos.

Referencias:
- https://www.listendata.com/2015/03/weight-of-evidence-woe-and-information.html
- https://medium.com/mlearning-ai/weight-of-evidence-woe-and-information-value-iv-how-to-use-it-in-eda-and-model-building-3b3b98efe0e8#:~:text=4-,Weight%20of%20Evidence%20(WoE)%20and%20Information%20Value%20(IV),distribution%20of%20good%20or%20bad.
- https://datasciencestunt.com/weight-of-evidence-woe-and-information-value-iv/

```python
# Definimos una función que nos permita cálculal el information value para cada variable
def information_value(feature, target, data):
    """information value

    Args:
        feature : variable a analizar del conjunto de features
        target (_type_): variable de target
        data (_type_): conjunto de datos
    """
    contingency_table = pd.crosstab(data[feature], data[target])
    event_total = contingency_table[0].sum()
    non_event_total = contingency_table[1].sum()
    
    woe_values = {}
    iv_values = {}
    
    for category in contingency_table.index:
        event_count = contingency_table.loc[category, 1]
        non_event_count = contingency_table.loc[category, 0]
        
        event_ratio = event_count / event_total
        non_event_ratio = non_event_count / non_event_total
        
        if event_ratio == 0:
            event_ratio += 0.0001  # Para evitar divisiones por cero, esto se le conoce cómo un método de censura del WOE ya que no es posible cálcular ln(0)
        if non_event_ratio == 0:
            non_event_ratio += 0.0001 
        
        woe = np.log(event_ratio / non_event_ratio)
        iv = (event_ratio - non_event_ratio) * woe
        
        woe_values[category] = woe
        iv_values[category] = iv

# Sumar los valores IV para obtener el IV total
    total_iv = sum(iv_values.values())
    
    return total_iv
    
```

```python
# Calculamos el information value para cada columna con un loop
iv_result = pd.DataFrame(columns=['valor'])
for column in df_features:
    iv_value = information_value(column, 'poisonous', df_full)
    result = pd.DataFrame({'valor':iv_value}, index = [column])
    iv_result = pd.concat([iv_result,result], ignore_index=False)
iv_result
```

Según cómo indica la teoría del information value, cuando tenemos resultados con un IV superior a 0.5 esto puede presentar un comportamiento sospechoco, en el sentido que puede ser una variable que puede sobreajustar el modelo.
```python
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

preprocessor_cat = make_pipeline(
        OneHotEncoder(drop='first')
        )
preprocessor_cat
```

## Definición y Entrenamiento de Modelos

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier


boosting_params = {
    'n_estimators': 200,
    'max_depth': 1,
    'learning_rate': 1,
    'criterion': 'squared_error'
}

estimators = {
    "svm": SVC(kernel="rbf",C=1,gamma="scale"),
    "decision_tree": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier(),
    "Random_Forest": RandomForestClassifier(n_estimators = 200, random_state = 123),
    "Boosting_Model": GradientBoostingClassifier(**boosting_params),
    "xgboost": XGBClassifier(verbosity = 0)
}

parametros = {
    "svm": [{
        "svm__kernel": ["rbf", "linear"],
        "svm__C": [0.1, 1, 10],
        "svm__gamma": ["scale", "auto"],
    }],
    "decision_tree": {
        "decision_tree__criterion": ["gini", "entropy"],
        "decision_tree__max_depth": [i for i in range(1,15)],
    },
    "KNN": {
        "KNN__n_neighbors": [i for i in range(1,20)],
        "KNN__weights": ["uniform", "distance"],
    },
    "Random_Forest": {
        'Random_Forest__n_estimators': [100, 200, 300],
        'Random_Forest__max_depth': [None, 10, 20],
        'Random_Forest__min_samples_split': [2, 5, 10],
    },
    "Boosting_Model": {
        'Boosting_Model__n_estimators': [100, 200, 300],
        'Boosting_Model__max_depth': [i for i in range(5,20,5)],
        'Boosting_Model__learning_rate': [0.1, 0.01, 0.001],
    },
    "xgboost": {
        "xgboost__n_estimators" : [100, 200, 300],
        "xgboost__max_depth" : [i for i in range(5,20,5)],
        "xgboost__learning_rate" : [0.1, 0.001, 0.0001],
        },
    
}
```

```python
from sklearn.model_selection import GridSearchCV

modelo = dict()
for estimador in estimators.keys():
    
    pipeline = Pipeline(steps = [
        ('Preprocesamiento', preprocessor_cat), 
        (estimador, estimators[estimador])]
                        )
    
    modelo[estimador] = GridSearchCV(pipeline, parametros[estimador], cv=5, scoring='accuracy')
    
    modelo[estimador].fit(X_train, y_train.values.ravel())
    print("Modelo {}".format(estimador))
    print("Los mejores parámetros para el estimador son {}".format( modelo[estimador].best_params_))
    print("Accuracy: {}".format(modelo[estimador].best_score_))
    print("El Score respecto a la base de prueba es: {}".format(modelo[estimador].score(X_test,y_test)))
    print("="*20)
```

```python
from sklearn.metrics import roc_curve, auc, f1_score, precision_score, recall_score, accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay
```
